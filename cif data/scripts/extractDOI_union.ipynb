```python
import numpy as np
from ccdc.io import CrystalReader
from ccdc.descriptors import CrystalDescriptors
from ccdc.diagram import DiagramGenerator
from ccdc.io import EntryReader
from ccdc import io
import pandas as pd
import pickle
```

```python
with open('DOI_REF.pickle', 'rb') as handle:
    doi_to_code = pickle.load(handle)
```

```python
df = pd.read_csv('CoRE_or_QMOF.csv')
df
```

```python
def getDOIJournal(refcode):
    with EntryReader('CSD') as reader:
        entry = reader.entry(refcode)
    
    return entry.publication.doi, entry.publication.journal
```

```python
DOI, journalName = [], []

for code in df['CSD code']:
    try:
        doi, journal = getDOIJournal(code)
        DOI.append(doi)
        journalName.append(journal)
    except:
        #this means that it's not a 6 letter ref code and thus, can't query this into CCDC API to get DOI and journal name
        doi = 'Not a valid ref code, thus does not exist.'
        journal = 'Not a valid ref code, thus does not exist.'
        DOI.append(doi)
        journalName.append(journal)
```

```python
len(DOI), len(journalName)
```

```python
df['DOI'] = DOI
df['Journal'] = journalName
```

```python
df.to_csv('CoRE_or_QMOF_properties.csv', index = False)
```
