```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('CoRE_or_QMOF_properties.csv')
df
```

```python
#get metal types
with open('CSD code map/metals/int_props.pickle', 'rb') as handle:
    intersection_metals = pickle.load(handle)

with open('CSD code map/metals/core_props.pickle', 'rb') as handle:
    core_metals = pickle.load(handle)

with open('CSD code map/metals/qmof_props.pickle', 'rb') as handle:
    qmof_metals = pickle.load(handle)
```

```python
len(intersection_metals), len(core_metals), len(qmof_metals)
```

```python
combined_dict = intersection_metals | core_metals | qmof_metals
len(combined_dict)
```

```python
metal_type_cumul = []

for id in df['CSD code']:
    metalList = combined_dict[id]
    metalString = ', '.join(metalList)
    metal_type_cumul.append(metalString)
```

```python
df['Metal types'] = metal_type_cumul
```

```python
df
```

```python
#get other properties
with open('CSD code map/other properties/int_props.pickle', 'rb') as handle:
    int_props = pickle.load(handle)

with open('CSD code map/other properties/core_props.pickle', 'rb') as handle:
    core_props = pickle.load(handle)

with open('CSD code map/other properties/qmof_props.pickle', 'rb') as handle:
    qmof_props = pickle.load(handle)
```

```python
int_props #[MOFid, smiles_nodes, smiles_linker, smiles, formula, space_group, crys_sys]
```

```python
combined_props = int_props | core_props | qmof_props
len(combined_props)
```

```python
mofid_cum, smiles_linker_cum, molecular_formula_cum, space_group_cum, crys_sys_cum = [], [], [], [], []

for id in df['CSD code']:
    listofprops = combined_props[id]
    
    try:
        mofid_cum.append(listofprops[0])
        smiles_linker_cum.append(listofprops[2])
        molecular_formula_cum.append(listofprops[4])
        space_group_cum.append(listofprops[5])
        crys_sys_cum.append(listofprops[6])
    except:
        mofid_cum.append(np.nan)
        smiles_linker_cum.append(np.nan)
        molecular_formula_cum.append(np.nan)
        space_group_cum.append(np.nan)
        crys_sys_cum.append(np.nan)
```

```python
len(mofid_cum), len(smiles_linker_cum), len(molecular_formula_cum), len(space_group_cum), len(crys_sys_cum)
```

```python
df['MOFid'] = mofid_cum
df['smiles_linker'] = smiles_linker_cum
df['Molecular formula'] = molecular_formula_cum
df['Space group'] = space_group_cum
df['Crystal system'] = crys_sys_cum
```

```python
df
```

```python
df.to_csv('CoRE_QMOF_expanded.csv', index = False)
```
